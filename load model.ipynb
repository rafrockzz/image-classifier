# 1. Load Model

```python
from tensorflow.keras.models import load_model

# Load the model
model = load_model('./models/imageclassifier.h5')
```

# 2. Specific Image Test

```python
import cv2
import os
import tensorflow as tf
import numpy as np
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('./data/covid/spectrograph_6.png')
plt.imshow(img)
plt.show()
```

```python
resize = tf.image.resize(img, (260,370))
plt.imshow(resize.numpy().astype(int))
plt.show()
```

```python
yhat = model.predict(np.expand_dims(resize, 0))
```

```python
yhat
```

```python
if yhat > 0.5: 
    print(f'Predicted class is healthy')
else:
    print(f'Predicted class is covid')
```

# 3. Multiple Images Test

```python
# Specify the folder containing the test images
test_folder = './data/covid/'

# Iterate through all the images in the folder
for filename in os.listdir(test_folder):
    # Load and preprocess the image
    img_path = os.path.join(test_folder, filename)
    img = cv2.imread(img_path)
    resize = cv2.resize(img, (370, 260))  # Resize the image as needed
    
    # Perform prediction
    yhat = model.predict(np.expand_dims(resize, 0))
    
    # Print the prediction result
    if yhat > 0.5: 
        print(f'Image {filename}: Predicted class is healthy', yhat)
    else:
        print(f'Image {filename}: Predicted class is COVIDDDDDDD', yhat)
```
